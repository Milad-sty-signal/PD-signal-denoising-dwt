```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file (update 'your_file.csv' to your actual filename)
df = pd.read_csv(r'C:\Users\Milad\Desktop\plots of paper\experimental TDR.csv')
```

```python
# === Multiply Columns B and C by 100 ===
df.iloc[:, 1] *= 100   # Column B
```

```python
# Add 1.057 to column A (Time)
#df.iloc[:, 0] += 19.5
```

```python
# Rename columns for clarity (optional but useful)
df.columns = ['Time', 'PD_Amplitude', 'noisy']
```

```python
df_filtered = df[(df['Time'] >= -0.100) & (df['Time'] <= 0.140)]
df_filtered = df_filtered.dropna(subset=['PD_Amplitude'])
```

```python
# === Plot ===
plt.figure(figsize=(3.5, 2.5), dpi=300)
plt.plot(df_filtered['Time'], df_filtered['PD_Amplitude'], color='black', linewidth=0.8)

# === Axis settings ===
plt.xlabel('Time [us]', fontsize=8)
plt.ylabel('Measured PD Amplitude [pC]', fontsize=8)
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)
plt.grid(True, linestyle='--', linewidth=0.3)
plt.tight_layout()
plt.ylim(-170, 80)

# Optional: save figure
plt.savefig('TDR of EX_test.tif', format='tiff', dpi=600)
```
